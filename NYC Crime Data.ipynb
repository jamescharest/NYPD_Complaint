## Cleaning and adding data to our dataframe

```python
df.info()
```

```python
df.columns
```

#### Add a "MONTH" column to our dataframe

```python
df["CMPLNT_FR_DT"] =  pd.to_datetime(df["CMPLNT_FR_DT"], format="%Y-%m-%d")
```

#### Above showed that the CMPLNT_FR_DT was not in datetime format

```python
df["MONTH"] = df["CMPLNT_FR_DT"].dt.month
```

```python
df.head()
```

```python
df["MONTH"].value_counts()
```

#### We already have a "DAY_OF_WK" Column, but testing to see how to add a "DAY" column to our dataframe

```python
df["DAY"] = df["CMPLNT_FR_DT"].dt.day_name()
```

```python
df.head()
```

#### Now we are going to see if we can turn the "MONTH" column to read the actual month

```python
import calendar
```

```python
df['MONTH'] = df['MONTH'].apply(lambda x: calendar.month_abbr[x])
```

```python
df.head()
```

```python
df["MONTH"].value_counts()
```

#### Eliminating the Final "DAY" column, because my cleaned data included that

```python
df.drop("DAY", inplace=True, axis=1)
```

```python
df.head()
```
